```python
import pandas as pd
import numpy  as np
```

```python
train_labels = pd.read_csv("data/train_original.csv")
labels       = pd.read_csv("data/labels.csv")
```

```python
# make df which express cell_type as one-hot
col            = ["id","cell_type"]
labels.columns = col
i              = pd.DataFrame([[0,"Nucleoplasm"]])
i.columns      = col
labels_df      = pd.concat([i,labels])

length          = len(labels_df)
len_array       = list(range(length))
labels_df.index = len_array
```

```python
# new df 
new_labels                = labels_df["cell_type"]
length                    = len(train_labels)
train_labels_edit         = pd.DataFrame([[0]*28]*length)
train_labels_edit.columns = new_labels
train_labels_edit.index   = train_labels["Id"]

# search id and modify value.
for pic_id, target in zip(train_labels["Id"],train_labels["Target"]):
    target_array_str = target.split(" ")
    target_array_int = [int(i) for i in target_array_str]
    for cell_id_int in target_array_int:
        cell_type_name = labels_df.at[cell_id_int, "cell_type"]
        train_labels_edit.at[pic_id, cell_type_name] = 1
      
display(train_labels_edit)
```

```python
train_labels_edit.to_csv("data/train.csv")
```
